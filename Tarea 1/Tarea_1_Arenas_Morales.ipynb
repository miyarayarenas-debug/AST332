#**0. Breve descripción del contexto de los datos**

Este conjunto de datos de canciones Spotify contiene un rango de 125 géneros diferentes, cada canción tiene alguna característica de audio asociada a ella. Estos datos están en formato '.csv' que corresponde a un formato tabular.

Cada columna de este dataset contiene lo siguiente:


*   **track_id** : El ID de Spotify para la canción.
*   **artist** : Los nombres de los artistas que interpretaron la canción. Si hay más de un artista, están separados por un ;
* **album_name** : El nombre del álbum en el que aparece la canción
* **track.name** : Nombre de la pista
* **popularity** : La popularidad de una pista es un valor entre 0 y 100, siendo 100 los más populares. La popularidad se calcula por algoritmo y se basa, en su mayor parte, en el número total de obras que ha tenido la pista y lo reciente que son esas obras. En términos generales, las canciones que se están tocando mucho ahora tendrán una mayor popularidad que las canciones que se tocaron mucho en el pasado. Las pistas duplicadas (por ejemplo, la misma pista de un solo y un álbum) se clasifican independientemente. La popularidad artística y discográfica se deriva matemáticamente de la popularidad de la pista.
* **duration_ms**: la longitud de la pista en milisegundos
* **explicit** : Si la pista tiene letras explícitas o no (verdadera = sí lo hace; false = no no es lo desconocido)
* **danceability**: La danza describe lo adecuada que es una pista para bailar a partir de una combinación de elementos musicales que incluyen tempo, estabilidad del ritmo, fuerza de ritmo y regularidad general. Un valor de 0.0 es menos bailable y 1.0 es el más bailable
* **energy**: La energía es una medida de 0.0 a 1.0 y representa una medida perceptual de intensidad y actividad. Normalmente, las pistas energéticas se sienten rápidas, ruidosas y ruidosas. Por ejemplo, Death Metal tiene alta energía, mientras que un preludio de Bach obtiene bajas en la escala
* **key**: La tecla de la pista está en. Los enteros mapean a los lanzamientos usando notación estándar de clase de tono. Por ejemplo. 0 = C, 1 = C♯/D♭, 2 = D, y así sucesivamente. Si no se detectó ninguna llave, el valor es -1
* **loudness**: El ruido general de una pista en decibelios (dB)
* **mode** : Modo indica la modalidad (mayor o menor) de una pista, el tipo de escala de la que se deriva su contenido melódico. Mayor está representado por 1 y menor es 0
* **speechiness** : Speechiness detecta la presencia de palabras habladas en una pista. La más exclusivamente habla-como la grabación (por ejemplo, talk show, audioliario, poesía), más cerca de 1.0 el valor de atributo. Valores por encima de 0,66 describen pistas que probablemente se hacen enteramente de palabras habladas. Los valores entre 0.33 y 0,66 describen temas que pueden contener tanto música como en el habla, ya sea en secciones o en capas, incluyendo casos como música rap. Los valores por debajo de 0.33 probablemente representan la música y otras pistas no-heechizadas
* **acousticness**: Una medida de confianza de 0.0 a 1.0 de si la pista es acústica. 1.0 representa una alta confianza en la que la pista es acústica
* **instrumentalness** : Predece si una canción no contiene voces. Los sonidos de "Oh" y "aah" se tratan como instrumentales en este contexto. Las pistas de rap o palabras habladas son claramente "vocales". Cuanto más se acerque el valor de la instrumentalidad a 1.0, mayor es la probabilidad de que la pista no contenga contenido vocal
* **liveness** : Detecta la presencia de un público en la grabación. Los valores de vida más altos representan una mayor probabilidad de que la pista se realizó en vivo. Un valor por encima de 0.8 proporciona una fuerte probabilidad de que la pista esté en vivo
* **valence** : Una medida de 0.0 a 1.0 describiendo la positiva musical transmitida por una canción. Las pistas con alta valencia suenan más positivas (por ejemplo, feliz, alegre, eufórico), mientras que las pistas con baja valencia suenan más negativas (por ejemplo, triste, deprimido, enfadado)
* **tempo** : El ritmo estimado general de una pista en latidos por minuto (BPM). En terminología musical, el tempo es la velocidad o el ritmo de una pieza dada y deriva directamente de la duración media del ritmo
* **time_signature**: Una firma de tiempo estimada. La firma de tiempo (metro) es una convención notativa para especificar cuántos latidos hay en cada barra (o medida). La firma de tiempo va de 3 a 7 indicando firmas de tiempo de 3/4, a 7/4.
* **track_genre** : El género en el que pertenece la pista

# 2. Limpieza de datos


En esta sección, se llevará a cabo el proceso de **limpieza y preparación** del dataset para asegurar la calidad y consistencia de los datos antes de proceder con el análisis exploratorio.

### a) Conversión de unidades

En la sección anterior pudimos notar que la columna correspondiente al tiempo de duración de cada canción se encuentre en **milisegundos** y por lo tanto vamos a realizar una conversión a **minutos** para estandarizar las unidades y facilitar la interpretación de las variables temporales.

```python
# conversión de milisegundos a minutos
spotify['duration_m'] = spotify['duration_ms'] / 60000  # se agrega otra columna
```

```python
spotify.head() # Última columna corresponde a la duración en minutos
```

### b) Manejo de valores nulos o faltantes

A simple vista en la info del DataFrame es posible notar que hay 3 columnas en las que difiere el número de elementos, para poder obtener la cantidad total de elementos que faltan/nulos por columna ocupamos la siguiente función de pandas.

```python
spotify.isnull().sum() # Nos dice cuántos valores nulos hay en cada columna
```

Ahora que estan identificadas las columnas donde se encuentran los valores nulos los rellenamos con nombres que sirva para identificar que no tienen info.

```python
spotify['artists']=spotify['artists'].fillna('Desconocido')
spotify['album_name']=spotify['album_name'].fillna('Desconocido')
spotify['track_name']=spotify['track_name'].fillna('Desconocido')
```

Ocupamos nuevamente la función anterior para verificar si se efectuaron los cambios y ahora no existe ningún campo nulo

```python
spotify.isnull().sum()
```

### c) Detección y eliminación de datos duplicados

Ahora que ya no queda ningún campo vacío, buscaremos eliminar las canciones que estan duplicadas si es que existen. Para esto, utilizaremos nuevamente una función propia de los DataFrame en pandas que nos entrega la cantidad de items que se encuentran duplicados.

```python
dup = spotify.duplicated().sum() # Busca los duplicados por cada fila y los suma
print(dup)
```

```python
duplicados = spotify['track_id'].duplicated().sum()
print(duplicados)
```

Podemos notar que hay muchos elementos de la columna 'track_id' que están duplicados, para ver por qué sucede esto vamos a extraer el primer set de canciones duplicadas para corroborar si todos los elemnetos de las columnas se repiten.

```python
dup_track_ids = spotify[spotify.duplicated(subset=['track_id'], keep=False)]['track_id'].unique() # Obtener los track_id que están duplicados

if len(dup_track_ids) > 0:  # Toma el primer track_id duplicado para inspeccionar.
    sample_track_id = dup_track_ids[0]
    print(f"\nFilas para el track_id duplicado '{sample_track_id}':")
    print(spotify[spotify['track_id'] == sample_track_id])
```

Aquí podemos notar que cada canción tiene su track_id único pero estos están duplicados porque hay algunas canciones que se clasifican con más de un género por lo tanto se repite la canción para cada 'track_genre' distinto. Debemos limpiar estos datos para que no se repitan los elementos por lo tanto vamos a agrupar cada género asociado a la canción en una lista de elementos dentro de la columna 'track_genre' para no perder esta información que es relevante.

```python
# Apéndice
aggregation_rules = {             # Para el resto de columnas consideramos la primera aparición
    'artists': 'first',           # ya que comprobamos que se repite todo, exepto 'track_genre'
    'album_name': 'first',
    'track_name': 'first',
    'popularity': 'first',
    'duration_ms': 'first',
    'explicit': 'first',
    'danceability': 'first',
    'energy': 'first',
    'key': 'first',
    'loudness': 'first',
    'mode': 'first',
    'speechiness': 'first',
    'acousticness': 'first',
    'instrumentalness': 'first',
    'liveness': 'first',
    'valence': 'first',
    'tempo': 'first',
    'time_signature': 'first',
    'duration_m': 'first',
    'track_genre': lambda x: list(x.unique()),  # Consolidar todos los géneros únicos en una lista
}

spotify_cleaned = spotify.groupby('track_id').agg(aggregation_rules).reset_index()
```

El código anterior agrupa el DataFrame por 'track_id' y aplica las reglas de agregación y esto creará un nuevo DataFrame donde cada 'track_id' es único y la columna 'track_genre' contendrá una lista de todos los géneros asociados a esa canción.

```python
spotify_cleaned.head()
```

Podemos notar que la cantidad de elementos disminuyó, en un principio teníamos **114.000** elementos pero al quitar los **24.259** duplicados nos quedamos con **89.741** datos limpios tal como se muestra a continuación.

```python
spotify_cleaned.info()
spotify_cleaned.describe()
```

Comparación con el mismo 'track_id' seleccionado anteriormente y se puede ver que tiene todos los géneros agrupados en una lista en la columna de 'track_genre'.

## Relación Valence vs. Danceability

Para visualizar esta relación de si las canciones consideradas más alegres son las más bailables se realizará un grafico tipo scatter en donde vamos a considerar solo los géneros que tengan mayor popularidad en promedio, esto se hace con el fin de acotar la gran cantidad de datos de nuestro DataFrame.

```python
sp_exploded = spotify.explode('track_genre') # Separa las listas en filas independientes
pop_by_genre = sp_exploded.groupby('track_genre')['popularity'].mean().reset_index() # Calculamos la popularidad promedio por género
top3_genres = pop_by_genre.nlargest(3, 'popularity')  # Obtenemos los 3 géneros con mayor popularidad promedio
top3_data = sp_exploded[sp_exploded['track_genre'].isin(top3_genres['track_genre'])]

sns.scatterplot(x='valence', y='danceability', hue = 'track_genre', data=top3_data,)  # Generamos el scatterplot para visualizar relaciones
plt.title('Relación entre valence y danceability en el Top 3 de géneros más populares')
plt.xlabel('Valence')
plt.ylabel('Danceability')
plt.show()
```

Gracias al scatter plot podemos decir que la hipotesis con la que se realizó este gráfico estaba erronea, ya que si bien existe una correlación entre las valencia y la bailabilidad, esta no es tan directa como se penso en algún momento. ## Matriz de Correlación

Ahora vamos a definir un nuevo DataFrame que solo contenga las variables numéricas haciendo una copia a 'spotify_cleaned'.

```python
num_cols = spotify_cleaned[spotify_cleaned.columns[(spotify_cleaned.dtypes == 'float64') | (spotify_cleaned.dtypes == 'int64')]].copy() # Copia del DataFrame solo con las columnas numéricas
num_cols.drop('duration_ms', axis = 1, inplace = True)  # Eliminamos la columna del tiempo en milisegundos
num_cols.shape
```

A continuación nos vamos a enfocar en cuantificar la relación entre las variables numéricas de nuestro DataFrame, esto lo vamos a realizar mediante la construcción de la **Matriz de Correlación** que nos indica cómo cada par de variables se relaciona entre sí. Esta correlación puede ser positiva (cercana a 1), negativa (cercana a -1) o bien puede ser nula (igual a 0).

```python
corr = num_cols.corr() # Matriz de correlación
mask = np.zeros_like(corr) # Mapa booleano que indica qué celdas del heatmap deben ocultarse.
mask[np.triu_indices_from(mask)] = True # Nos oculta el triángulo superior de la matriz.
cmap = sns.diverging_palette(145, 300, s=60, as_cmap=True)
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(14, 10))
    sns.heatmap(
        corr,
        mask=mask,
        vmax=1,
        vmin = -1,
        square=True,
        cmap = cmap,
        annot = True,
        annot_kws={"size":10, "fontname":'DejaVu Sans Mono'},
        )
    ax.set_title('Matriz de correlación', fontname='DejaVu Sans Mono', size=12)
    plt.show()
```
